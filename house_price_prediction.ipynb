#HOUSE_PRICE_PRIDICTION

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, LabelEncoder
df = None
def load_data():
    global df
    df = pd.read_csv("data.csv")
    print("Data loaded successfully!\n")
    print(df)
def show_info():
    if df is not None:
        print(df.head())
        print(df.info())
        print(df.isnull().sum())
    else:
        print("Load data first!\n")
def clean_columns():
    global df
    if df is not None:
        df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
        print("Column names cleaned!\n")
    else:
        print("Load data first!\n")
print(df)
def handle_missing():
    global df
    if df is not None:
        df.fillna(df.median(numeric_only=True), inplace=True)

        for col in df.select_dtypes(include='object'):
            df[col].fillna(df[col].mode()[0], inplace=True)

        print("Missing values handled!\n")
    else:
        print("Load data first!\n")
def remove_duplicates():
    global df
    if df is not None:
        df.drop_duplicates(inplace=True)
        print("Duplicates removed!\n")
    else:
        print("Load data first!\n")
def remove_outliers():
    global df
    if df is not None:
        for col in df.select_dtypes(include=np.number):
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1
            df = df[(df[col] >= Q1 - 1.5 * IQR) &
                    (df[col] <= Q3 + 1.5 * IQR)]
        print("Outliers removed!\n")
    else:
        print("Load data first!\n")
def preprocess_data():
    global df
    if df is not None:
        scaler = StandardScaler()
        num_cols = df.select_dtypes(include=np.number).columns
        df[num_cols] = scaler.fit_transform(df[num_cols])

        le = LabelEncoder()
        for col in df.select_dtypes(include='object'):
            df[col] = le.fit_transform(df[col])

        print("Data preprocessing completed!\n")
    else:
        print("Load data first!\n")
def skew_kurtosis():
    global df
    if df is not None:
        print("Skewness:\n", df.skew(numeric_only=True))
        print("\nKurtosis:\n", df.kurt(numeric_only=True))
    else:
        print("Load data first!\n")
def probability_distribution():
    global df
    if df is not None:
        num_cols = df.select_dtypes(include=np.number).columns
        for col in num_cols:
            data = df[col].dropna()
            if len(data) == 0:
                continue
            plt.figure()
            plt.hist(data, bins=30, density=True)
            try:
                data.plot(kind='kde')
            except:
                print(f"KDE not available for {col} (scipy missing)")
            plt.title(f"Probability Distribution of {col}")
            plt.xlabel(col)
            plt.ylabel("Density")
            plt.show()
    else:
        print("Load data first!\n")
def visualize():
    if df is not None:
        df.hist(figsize=(10, 8))
        plt.show()
    else:
        print("Load data first!\n")
def save_data():
    if df is not None:
        df.to_csv("cleaned_data.csv", index=False)
        print("Data saved successfully!\n")
    else:
        print("Load data first!\n")
while True:
    print("===== DATA CLEANING MENU =====")
    print("1. Load Data")
    print("2. Show Data Info")
    print("3. Clean Column Names")
    print("4. Handle Missing Values")
    print("5. Remove Duplicates")
    print("6. Remove Outliers")
    print("7. Data preprocess data")
    print("8.  find skew kurtosis")
    print("9. probability distribution")
    print("10. Visualize Data")
    print("11. Save Cleaned Data")
    print("12. Exit")
    choice = input("Enter your choice: ")
    if choice == "1":
        load_data()
    elif choice == "2":
        show_info()
    elif choice == "3":
        clean_columns()
    elif choice == "4":
        handle_missing()
    elif choice == "5":
        remove_duplicates()
    elif choice == "6":
        remove_outliers()
    elif choice == "7":
        preprocess_data()
    elif choice == "8":
        skew_kurtosis()
    elif choice == "9":
        probability_distribution()
    elif choice == "10":
        visualize()
    elif choice == "11":
        save_data()
    elif choice == "12":
        print("Exiting program...")
        break
    else:
        print("Invalid choice! Try again.\n")
```
